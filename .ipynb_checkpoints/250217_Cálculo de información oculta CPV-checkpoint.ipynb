```python
# 2. HOMOLOGACIÓN DE ID URBANA - CVEGEO
def homologate_urban_id(row):
    return f"{row['ENTIDAD']}{row['MUN']}{row['LOC']}{row['AGEB']}{row['MZA']}"

MZNA_2020_df['CVEGEO'] = MZNA_2020_df.apply(homologate_urban_id, axis=1)
MZNA_2010_df['CVEGEO'] = MZNA_2010_df.apply(homologate_urban_id, axis=1)
```

```python
# 3. FUNCIÓN PARA CONTAR * EN CAMPOS CLAVE
def calcular_porcentaje_asteriscos(df, campos):
    total_filas = len(df)
    resumen = {}
    
    for campo in campos:
        cantidad_asteriscos = df[campo].astype(str).str.contains(r'\*', na=False).sum()
        porcentaje = (cantidad_asteriscos / total_filas) * 100 if total_filas > 0 else 0
        resumen[campo] = {'Cantidad': cantidad_asteriscos, 'Porcentaje': porcentaje}
    
    return resumen

campos_interes = ['POBTOT', 'VIVTOT', 'VIVPAR_DES']

asteriscos_2020 = calcular_porcentaje_asteriscos(MZNA_2020_df, campos_interes)
asteriscos_2010 = calcular_porcentaje_asteriscos(MZNA_2010_df, campos_interes)

print("Resumen de asteriscos en 2020:", asteriscos_2020)
print("Resumen de asteriscos en 2010:", asteriscos_2010)
```

```python
# 2.3 Guardado de información en CSV
def save_to_csv(dataframe, filename):
    dataframe.to_csv(filename, index=False)

save_to_csv(MZNA_2020_df, MZNA_2020_salida)
save_to_csv(MZNA_2010_df, MZNA_2010_salida)
```

```python
# 4. FILTRADO POR MUNICIPIOS PERTENECIENTES A METRÓPOLIS
CVE_METRO_ruta = r"C:\Users\quesa\Desktop\Ejercicio\ClavesMetro.csv"
CVE_METRO_df = pd.read_csv(CVE_METRO_ruta, dtype={'CVE_MET': str, 'CVEMUN': str})

MZNA_2020_df['CVEMUN'] = MZNA_2020_df['ENTIDAD'].astype(str) + MZNA_2020_df['MUN'].astype(str)
MZNA_2010_df['CVEMUN'] = MZNA_2010_df['ENTIDAD'].astype(str) + MZNA_2010_df['MUN'].astype(str)

valores_filtro = CVE_METRO_df['CVEMUN'].unique()
MZNA_2020_df = MZNA_2020_df[MZNA_2020_df['CVEMUN'].isin(valores_filtro)]
MZNA_2010_df = MZNA_2010_df[MZNA_2010_df['CVEMUN'].isin(valores_filtro)]
```

```python
# 5. CÁLCULO DE PORCENTAJE DE * POR METRÓPOLI
MZNA_2020_df = MZNA_2020_df.merge(CVE_METRO_df[['CVEMUN', 'CVE_MET', 'NOM_MET']], on='CVEMUN', how='left')
MZNA_2010_df = MZNA_2010_df.merge(CVE_METRO_df[['CVEMUN', 'CVE_MET', 'NOM_MET']], on='CVEMUN', how='left')

def calcular_porcentaje_por_metro(df, campos):
    resultados = {}
    
    for campo in campos:
        total_por_metro = df.groupby('NOM_MET')[campo].count()
        con_asterisco = df[df[campo].astype(str).str.contains(r'\*', na=False)].groupby('NOM_MET')[campo].count()
        
        porcentaje_por_metro = (con_asterisco / total_por_metro * 100).fillna(0)
        resultados[campo] = porcentaje_por_metro
    
    return pd.DataFrame(resultados)

asteriscos_por_metro_2020 = calcular_porcentaje_por_metro(MZNA_2020_df, campos_interes)
asteriscos_por_metro_2010 = calcular_porcentaje_por_metro(MZNA_2010_df, campos_interes)

print("Porcentaje de * por metrópoli en 2020:\n", asteriscos_por_metro_2020)
print("Porcentaje de * por metrópoli en 2010:\n", asteriscos_por_metro_2010)
```

```python
# 6. GUARDADO DE RESULTADOS EN CSV
asteriscos_por_metro_2020.to_csv("Porcentaje_Asteriscos_2020.csv")
asteriscos_por_metro_2010.to_csv("Porcentaje_Asteriscos_2010.csv")
```

```python
# Asegurar que los campos relevantes son numéricos
for columna in ['POBTOT', 'VIVTOT', 'VIVPAR_DES', 'TVIVHAB']:
    MZNA_2020_df[columna] = pd.to_numeric(MZNA_2020_df[columna], errors='coerce')
    MZNA_2010_df[columna] = pd.to_numeric(MZNA_2010_df[columna], errors='coerce')

# Filtrar solo las manzanas (MZA ≠ "000") y municipios (LOC == "0000")
manzanas_2020 = MZNA_2020_df[MZNA_2020_df['MZA'] != "000"]
municipios_2020 = MZNA_2020_df[MZNA_2020_df['LOC'] == "0000"]

manzanas_2010 = MZNA_2010_df[MZNA_2010_df['MZA'] != "000"]
municipios_2010 = MZNA_2010_df[MZNA_2010_df['LOC'] == "0000"]

# Agrupar por metrópoli y sumar los campos requeridos
suma_manzanas_2020 = manzanas_2020.groupby('NOM_MET')[['POBTOT', 'VIVTOT', 'VIVPAR_DES', 'TVIVHAB']].sum()
suma_municipios_2020 = municipios_2020.groupby('NOM_MET')[['POBTOT', 'VIVTOT', 'VIVPAR_DES', 'TVIVHAB']].sum()

suma_manzanas_2010 = manzanas_2010.groupby('NOM_MET')[['POBTOT', 'VIVTOT', 'VIVPAR_DES', 'TVIVHAB']].sum()
suma_municipios_2010 = municipios_2010.groupby('NOM_MET')[['POBTOT', 'VIVTOT', 'VIVPAR_DES', 'TVIVHAB']].sum()

# Función para calcular la diferencia y el porcentaje
def calcular_diferencia_y_porcentaje(suma_manzanas, suma_municipios):
    diferencia = suma_municipios - suma_manzanas
    porcentaje = (diferencia / suma_municipios) * 100
    
    resultado = pd.DataFrame({
        'Suma Manzanas POBTOT': suma_manzanas['POBTOT'],
        'Suma Total Municipal POBTOT': suma_municipios['POBTOT'],
        'Diferencia POBTOT': diferencia['POBTOT'],
        'Porcentaje Diferencia POBTOT': porcentaje['POBTOT'],
        
        'Suma Manzanas VIVTOT': suma_manzanas['VIVTOT'],
        'Suma Total Municipal VIVTOT': suma_municipios['VIVTOT'],
        'Diferencia VIVTOT': diferencia['VIVTOT'],
        'Porcentaje Diferencia VIVTOT': porcentaje['VIVTOT'],

        'Suma Manzanas VIVPAR_DES': suma_manzanas['VIVPAR_DES'],
        'Suma Total Municipal VIVPAR_DES': suma_municipios['VIVPAR_DES'],
        'Diferencia VIVPAR_DES': diferencia['VIVPAR_DES'],
        'Porcentaje Diferencia VIVPAR_DES': porcentaje['VIVPAR_DES'],

        'Suma Manzanas TVIVHAB': suma_manzanas['TVIVHAB'],
        'Suma Total Municipal TVIVHAB': suma_municipios['TVIVHAB'],
        'Diferencia TVIVHAB': diferencia['TVIVHAB'],
        'Porcentaje Diferencia TVIVHAB': porcentaje['TVIVHAB']
    })

    return resultado

# Aplicar la función para 2020 y 2010
diferencia_2020 = calcular_diferencia_y_porcentaje(suma_manzanas_2020, suma_municipios_2020)
diferencia_2010 = calcular_diferencia_y_porcentaje(suma_manzanas_2010, suma_municipios_2010)

# Mostrar resultados
print("Diferencias 2020")
print(diferencia_2020)
print("\nDiferencias 2010")
print(diferencia_2010)
```
