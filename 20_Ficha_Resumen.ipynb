## GrÃ¡ficas

```python
#normal
def graficar_datos_resaltados(gdf, columna, cve_met_interes, output_path=None):
    # Extraer datos de la columna
    datos = gdf[columna].sort_values(ascending=False).reset_index(drop=True)
    media = datos.mean()
    
    # Identificar la posiciÃ³n de la metrÃ³poli de interÃ©s
    valor_interes = gdf.loc[gdf['CVE_MET'] == cve_met_interes, columna].values[0]
    posicion_interes = datos[datos == valor_interes].index[0]
    
    # Crear la grÃ¡fica
    fig, ax = plt.subplots(figsize=(5.31, 3.06))  # Ajustar las proporciones
    ax.plot(range(len(datos)), datos, color="darkgray",linewidth=1.5, label="Datos")
    
    # Resaltar la posiciÃ³n de interÃ©s
    ax.scatter([posicion_interes], [valor_interes], color="blue", label=f'{cve_met_interes}', zorder=5)
    
    # LÃ­nea de la media
    ax.axhline(y=media, color="red", linestyle="--", linewidth=1.5, label="Media metropolitana")

    # Invertir el eje X
    ax.invert_xaxis()
    
    # Estilo
#    ax.set_title("Comparativa de Datos", fontsize=12, fontname="Century Gothic")
    ax.set_xlabel("PosiciÃ³n Nacional", fontsize=8, fontname="Century Gothic")
    ax.set_ylabel(columna, fontsize=8, fontname="Century Gothic")
    ax.tick_params(axis="both", labelsize=7)
    ax.legend(fontsize=8, loc="upper left")
    ax.set_xlim([len(datos) - 1, 0])
    
    # Guardar la grÃ¡fica si se especifica
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.close(fig)

    # if output_path:
    #     plt.savefig(output_path, dpi=300, bbox_inches="tight")
    # else:
    #     plt.show()




#Invertida
def graficar_datos_invertido(gdf, columna, cve_met_interes, output_path=None):
    # Extraer datos de la columna
    datos = gdf[columna].sort_values().reset_index(drop=True)
    media = datos.mean()

    # Identificar la posiciÃ³n de la metrÃ³poli de interÃ©s
    valor_interes = gdf.loc[gdf['CVE_MET'] == cve_met_interes, columna].values[0]
    posicion_interes = datos[datos == valor_interes].index[0]
    
    # Crear la grÃ¡fica
    fig, ax = plt.subplots(figsize=(5.31, 3.06))  # Ajustar las proporciones
    ax.plot(range(len(datos)), datos, color="darkgray",linewidth=1.5, label="Datos")
    
    # Resaltar la posiciÃ³n de interÃ©s
    ax.scatter([posicion_interes], [valor_interes], color="blue", label=f'{cve_met_interes}', zorder=5)
    
    # LÃ­nea de la media
    ax.axhline(y=media, color="red", linestyle="--", linewidth=1.5, label="Media metropolitana")
    
    # Invertir ambos ejes
    ax.invert_yaxis()
    ax.invert_xaxis()
    
    # Estilo
#    ax.set_title("Comparativa de Datos", fontsize=12, fontname="Century Gothic")
    ax.set_xlabel("PosiciÃ³n Nacional", fontsize=8, fontname="Century Gothic")
    ax.set_ylabel(columna, fontsize=8, fontname="Century Gothic")
    ax.tick_params(axis="both", labelsize=7)
    ax.legend(fontsize=8, loc="upper left")
    ax.set_xlim([len(datos) - 1, 0])
    
    # Guardar la grÃ¡fica si se especifica
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.close(fig)

    # if output_path:
    #     plt.savefig(output_path, dpi=300, bbox_inches="tight")
    # else:
    #     plt.show()
```

## Exportar fichas

```python
import os
import pandas as pd
import win32com.client as win32

# Ruta de entrada y salida
input_folder = r"Z:\VOCES\DATA\Tratados\NAC\FICHAS\Resultados"
output_folder = r"Z:\VOCES\DATA\Tratados\NAC\FICHAS\PDFs"
os.makedirs(output_folder, exist_ok=True)

# Itera sobre el dataframe
for index, row in df_loc.iterrows():
    cve_met = row["CVE_MET"]
    file_path = os.path.join(input_folder, f"{cve_met}.xlsx")
    output_pdf = os.path.join(output_folder, f"{cve_met}.pdf")

    # Verifica que el archivo exista
    if not os.path.exists(file_path):
        print(f"Archivo no encontrado: {file_path}")
        continue

    # Abre Excel con COM
    excel = win32.gencache.EnsureDispatch('Excel.Application')
    wb = excel.Workbooks.Open(file_path)

    try:
        # Exporta a PDF (puedes ajustar el Ã­ndice de la hoja o usar el nombre si es necesario)
        wb.ExportAsFixedFormat(0, output_pdf)  # 0 = PDF format
        print(f"{cve_met} exportado correctamente.")
    except Exception as e:
        print(f"Error al exportar {cve_met}: {e}")
    finally:
        wb.Close(False)

    excel.Quit()

print("ExportaciÃ³n a PDF completada.")
```
